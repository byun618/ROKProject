```python
from tkinter import *
from tkinter import messagebox
from tkinter import filedialog as fd
from PIL import ImageTk, Image
import glob
from enum import Enum
import os.path
import threading
import time
import numpy as np
import re
from google.cloud import vision
from google.cloud.vision import types
import cv2
import pandas as pd
```

```python
class Detect:
    
    def detect(self, path, name, power, kill, dead, support):

        char = self.detect_text(path)
        li = self.change_to_list(char)
        
        name.append(li[0])
        power.append(self.change_to_number(li[1]))
        kill.append(self.change_to_number(li[2]))
        dead.append(self.change_to_number(li[3]))
        support.append(self.change_to_number(li[4]))
        
        return name, power, kill, dead, support

    def detect_text(self, path):
        """Detects text in the file."""
        from google.cloud import vision
        import io
        client = vision.ImageAnnotatorClient()

        # [START vision_python_migration_text_detection]
        with io.open(path, 'rb') as image_file:
            content = image_file.read()

        image = vision.types.Image(content=content)

        response = client.text_detection(image=image)
        texts = response.text_annotations

        return texts[0].description

    def change_to_number(self, s_power):
        list_of_numbers = re.findall(r'\d+', s_power)
        return int(''.join(list_of_numbers))

    def change_to_list(self, chars):
        tmp_texts = list()
        s = ""
        for c in chars:
            if c != '\n':
                s += c
            else:
                tmp_texts.append(s)
                s = ""

        return tmp_texts
```

```python
class Crop:
    
    def crop_name(self, initial, fn, x1, y1, x2, y2):
        
        crop = initial.crop((x1, y1, x2, y2))
        
        img = np.array(crop) 
        img = img[:, :, ::-1].copy()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        structuringElement = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

        imgTopHat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, structuringElement)
        imgBlackHat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, structuringElement)

        imgGrayscalePlusTopHat = cv2.add(gray, imgTopHat)
        gray = cv2.subtract(imgGrayscalePlusTopHat, imgBlackHat)

        img_blurred = cv2.GaussianBlur(gray, ksize=(5, 5), sigmaX=0)

        ret, img_thresh = cv2.threshold(
            img_blurred, 
            210, #name : 180
            255,
            cv2.THRESH_BINARY
        )
        
        cv2.imwrite('data/member/croped/name/' + fn, img_thresh)
    
    def crop_power(self, initial, fn, x1, y1, x2, y2):
        
        crop = initial.crop((x1, y1, x2, y2))
        
        img = np.array(crop) 
        img = img[:, :, ::-1].copy()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        structuringElement = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

        imgTopHat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, structuringElement)
        imgBlackHat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, structuringElement)

        imgGrayscalePlusTopHat = cv2.add(gray, imgTopHat)
        gray = cv2.subtract(imgGrayscalePlusTopHat, imgBlackHat)

        img_blurred = cv2.GaussianBlur(gray, ksize=(5, 5), sigmaX=0)

        ret, img_thresh = cv2.threshold(
            img_blurred, 
            210, #name : 180
            255,
            cv2.THRESH_BINARY
        )
        
        cv2.imwrite('data/member/croped/power/' + fn, img_thresh)
    
    def crop_kill(self, initial, fn, x1, y1, x2, y2):
        
        crop = initial.crop((x1, y1, x2, y2))
        
        img = np.array(crop) 
        img = img[:, :, ::-1].copy()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        structuringElement = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

        imgTopHat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, structuringElement)
        imgBlackHat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, structuringElement)

        imgGrayscalePlusTopHat = cv2.add(gray, imgTopHat)
        gray = cv2.subtract(imgGrayscalePlusTopHat, imgBlackHat)

        img_blurred = cv2.GaussianBlur(gray, ksize=(5, 5), sigmaX=0)

        ret, img_thresh = cv2.threshold(
            img_blurred, 
            210, #name : 180
            255,
            cv2.THRESH_BINARY
        )
        
        cv2.imwrite('data/member/croped/kill/' + fn, img_thresh)
        
    def crop_dead(self, initial, fn, x1, y1, x2, y2):
        
        crop = initial.crop((x1, y1, x2, y2))
        
        img = np.array(crop) 
        img = img[:, :, ::-1].copy()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        structuringElement = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

        imgTopHat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, structuringElement)
        imgBlackHat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, structuringElement)

        imgGrayscalePlusTopHat = cv2.add(gray, imgTopHat)
        gray = cv2.subtract(imgGrayscalePlusTopHat, imgBlackHat)

        img_blurred = cv2.GaussianBlur(gray, ksize=(5, 5), sigmaX=0)

        ret, img_thresh = cv2.threshold(
            img_blurred, 
            190, #name : 180
            255,
            cv2.THRESH_BINARY
        )
        
        cv2.imwrite('data/member/croped/dead/' + fn, img_thresh)
        
    def crop_support(self, initial, fn, x1, y1, x2, y2):
        
        crop = initial.crop((x1, y1, x2, y2))
        
        img = np.array(crop) 
        img = img[:, :, ::-1].copy()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        structuringElement = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

        imgTopHat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, structuringElement)
        imgBlackHat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, structuringElement)

        imgGrayscalePlusTopHat = cv2.add(gray, imgTopHat)
        gray = cv2.subtract(imgGrayscalePlusTopHat, imgBlackHat)

        img_blurred = cv2.GaussianBlur(gray, ksize=(5, 5), sigmaX=0)

        ret, img_thresh = cv2.threshold(
            img_blurred, 
            190, #name : 180
            255,
            cv2.THRESH_BINARY
        )
        
        cv2.imwrite('data/member/croped/support/' + fn, img_thresh)
```

```python
class Main(Frame):
    def __init__(self, parent):
        self.parent = parent
        self.frame = Frame(self.parent)
        self.drag_area = None
        self.ratio = 0.5
        self.n = 0
        
        self.name = list()
        self.power = list()
        self.kill = list()
        self.dead = list()
        self.support = list()
        
        self.coord = list()
        
        btn_edit_img = Button(parent, text='이미지 편집', command=self.crop_image)
        btn_edit_img.grid(column=0, row=0, padx=5, pady=5)
        
        btn_detect_img = Button(parent, text='이미지 인식', command=self.detect_image)
        btn_detect_img.grid(column=1, row=0, padx=5, pady=5)
        
        btn_export_img = Button(parent, text='엑셀로', command=self.export)
        btn_export_img.grid(column=2, row=0, padx=5, pady=5)
        
        self.lbl_status = Label(parent, text='진행 상황')
        self.lbl_status.grid(column=3, row=0, padx=5, pady=5, columnspan=4)
        
        # Load Images
        self.images = glob.glob('data/member/original/*') 
        image = Image.open(self.images[0])
        
        # Resize images
        w, h = int(image.size[0] * self.ratio), int(image.size[1] * self.ratio)
        image = image.resize((w, h))
        
        # Canvas
        self.origin_canvas = Canvas(self.parent, width=w, height=h)
        self.origin_canvas.grid(column=1, row=1, padx=1, pady=5, columnspan=3)
        
        # Show images on origin_canvas
        img = ImageTk.PhotoImage(image)
        self.origin_canvas.create_image(0, 0, anchor=NW, image=img)
        self.origin_canvas.image = img
        
        messagebox.showinfo("영역 지정", "이름 영역 지정해주세요")
        
        self.origin_canvas.bind("<Button-1>", self.click)
        self.origin_canvas.bind("<B1-Motion>", self.draw)
        self.origin_canvas.bind("<ButtonRelease-1>", self.release)

    def crop_image(self):
        
        total = len(self.images)
        
        s = '{}/{} 완료'.format(self.n, total)
        self.lbl_status.config(text=s)
        self.parent.update()
        
        pt, fn = os.path.split(self.images[self.n])
        
        initial = Image.open(self.images[self.n])
        
        self.coord.append((268, 192, 511, 228)) 
        self.coord.append((563, 197, 735, 224))
        self.coord.append((795, 196, 946, 227))
        self.coord.append((755, 405, 924, 447)) 
        self.coord.append((745, 567, 928, 606))
        
        crop = Crop()
        crop.crop_name(initial, fn, self.coord[0][0]/self.ratio, self.coord[0][1]/self.ratio,  self.coord[0][2]/self.ratio,  self.coord[0][3]/self.ratio)
        crop.crop_power(initial, fn, self.coord[1][0]/self.ratio, self.coord[1][1]/self.ratio,  self.coord[1][2]/self.ratio,  self.coord[1][3]/self.ratio)
        crop.crop_kill(initial, fn, self.coord[2][0]/self.ratio, self.coord[2][1]/self.ratio,  self.coord[2][2]/self.ratio,  self.coord[2][3]/self.ratio)
        crop.crop_dead(initial, fn, self.coord[3][0]/self.ratio, self.coord[3][1]/self.ratio,  self.coord[3][2]/self.ratio,  self.coord[3][3]/self.ratio)
        crop.crop_support(initial, fn, self.coord[4][0]/self.ratio, self.coord[4][1]/self.ratio,  self.coord[4][2]/self.ratio,  self.coord[4][3]/self.ratio)
            
        self.paste_image(self.n)
    
        self.n += 1
        
        s = '{}/{} 완료'.format(self.n, total)
        self.lbl_status.config(text=s)
        self.parent.update()

        if self.n == total:
            self.parent.after_cancel(self.job_load) 
            self.n = 0
            messagebox.showinfo("작업 상태", "완료!")
        else:
            self.job_load = self.parent.after(0, self.crop_image)
            
    def paste_image(self, n):
        
        images = glob.glob('data/member/croped/name/*')
        path, file_name = os.path.split(images[n])
        w, h = (0, 0)

        name = Image.open('data/member/croped/name/' + file_name)
        if w < name.size[0]:
            w = name.size[0]
        h += name.size[1]

        power = Image.open('data/member/croped/power/' + file_name)
        if w < power.size[0]:
            w = power.size[0]
        h += power.size[1]

        kill = Image.open('data/member/croped/kill/' + file_name)
        if w < kill.size[0]:
            w = kill.size[0]
        h += kill.size[1]

        dead = Image.open('data/member/croped/dead/' + file_name)
        if w < dead.size[0]:
            w = dead.size[0]
        h += dead.size[1]

        support = Image.open('data/member/croped/support/' + file_name)
        if w < support.size[0]:
            w = support.size[0]
        h += support.size[1]

        new = Image.new("RGB", (w+50, h+50))
        new.paste(name, (0, 0))
        new.paste(power, (0, name.size[1]))
        new.paste(kill, (0, name.size[1]+power.size[1]))
        new.paste(dead, (0, name.size[1]+power.size[1]+kill.size[1]))
        new.paste(support, (0, name.size[1]+power.size[1]+kill.size[1]+dead.size[1]))

        new.save('data/member/pasted/' + file_name)
        
    def detect_image(self):
                
        total = len(self.images)
        
        s = '{}/{} 완료'.format(self.n, total)
        self.lbl_status.config(text=s)
        self.parent.update()
        
        detect = Detect()
        pt, fn = os.path.split(self.images[self.n])
        print(self.n, fn)
        self.name, self.power, self.kill, self.dead, self.support = detect.detect('data/member/pasted/' + fn, self.name, self.power, self.kill, self.dead, self.support)
    
        self.n += 1
        
        s = '{}/{} 완료'.format(self.n, total)
        self.lbl_status.config(text=s)
        self.parent.update()

        if self.n == total:
            self.parent.after_cancel(self.job_detect) 
            self.n = 0
            messagebox.showinfo("작업 상태", "완료!")
        else:
            self.job_detect = self.parent.after(0, self.detect_image)
            
            
    def export(self):
        df = pd.DataFrame({'name': self.name,
                            'power':self.power,
                            'kill':self.kill,
                            'dead':self.dead,
                            'support':self.support})
        
        filename = fd.asksaveasfilename(filetypes=[('excel file', 'xlsx')], title='Save file as', initialfile='memberinfo.xlsx')
        df.to_excel(filename)
        
        
    def click(self, event):
        self.old_x, self.old_y = event.x, event.y
        
        
    def draw(self, event):
        if self.drag_area != None:
            self.origin_canvas.delete(self.drag_area)
        
        w = 2
        self.drag_area = self.origin_canvas.create_rectangle(self.old_x, self.old_y, event.x, event.y, width=w)
        
    def release(self, event):
        self.coord.append((self.old_x, self.old_y, event.x, event.y))

        w = 2
        
        for i in range(len(self.coord)):
            self.origin_canvas.create_rectangle(self.coord[i][0], self.coord[i][1], self.coord[i][2], self.coord[i][3], width=w)
        
        if len(self.coord) == 1:
            messagebox.showinfo("영역 지정", "투력 영역 지정해주세요")
        elif len(self.coord) == 2:
            messagebox.showinfo("영역 지정", "투력 옆 처치 영역 지정해주세요")
        elif len(self.coord) == 3:
            messagebox.showinfo("영역 지정", "전사 영역 지정해주세요")
        elif len(self.coord) == 4:
            messagebox.showinfo("영역 지정", "원조 영역 지정해주세요")
        elif len(self.coord) == 5:
            messagebox.showinfo("영역 지정", "모든 영역 지정 완료")
            print(self.coord)
        
        
```
